## Argumentos Flexíveis `*args` `**kwargs`

São usados quando não se sabe quantos argumentos serão passados para a função pelo usuário

`` *args`` e `` **kwargs`` capturam todos os argumentos passados.

```python
def exibe_argumentos(*args, **kwargs):
    #return type(kwargs)
    print("Argumentos: ", args)    
    print("Argumentos de palavras-chaves: ", kwargs)  
```

```python
exibe_argumentos(10,50,'abc',param='valor')
```

Na verdade, os nomes `` args`` e `` kwargs`` não importam, mas os caracteres `` * `` que os precedem.

`` args`` e `` kwargs`` são apenas os nomes de variáveis frequentemente usados por convenção, abreviação de "argumentos" e "argumentos de palavras-chave".

um `` * `` antes de uma variável significa "expandir como uma tupla", enquanto um duplo `` ** `` antes de uma variável significa "expandir como um dicionário".

```python
def exibe_argumentos2(*asdasd, **a1):
    print("Argumentos: ", asdasd)    
    print("Argumentos de palavras-chaves: ", a1)  
    
exibe_argumentos2(10,20,30, p=50) 
```
