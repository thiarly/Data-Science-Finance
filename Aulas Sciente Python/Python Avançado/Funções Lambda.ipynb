## Qual a utilidade?

### Exemplo

A função ``sorted`` do Python ordena os element

```python
sorted([2,4,3,5,1,6])
```

Mas dicionários não são ordenáveis: precisamos de uma maneira de dizer à função `sorted` *como* ordernar nossos dados.

```python
ativos = [{'ticker':'VALE3', 'date':'2020-06-07', 'price':56.57},
          {'ticker':'ITSA4', 'date':'2020-06-07', 'price':10.36},
          {'ticker':'MGLU3', 'date':'2020-06-07', 'price':71.99}]
ativos
```

#### Ordenando por Ticker

```python
#ativos[2]['ticker']
sorted(ativos, key=lambda item: item['ticker'])
```

#### Ordenando por preço

```python
sorted(ativos, key=lambda item: item['price'])
```

```python
def ordenar(item):
    return item['price']
```

```python
sorted(ativos, key=ordenar)  
```
