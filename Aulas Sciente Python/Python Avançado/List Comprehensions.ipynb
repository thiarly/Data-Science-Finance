### forma tradiconal

```python
# Forma tradicional
quadrados = []
for numero in range(10):
  quadrados.append(numero**2)
quadrados
```

###List comprehension 
#### [expressão for variável in iterável]

```python
# List Comprehension
quad = [x**2 for x in range(10)]
quad
```

```python
dobro = [x*2 for x in [3,4,10]]
dobro
```

### Iterações Múltiplas
#### [expressão for variável in interável for variável2 in interável2]

```python
[(i, j) for i in range(2) for j in range(3)]
```

### Teste condicional
#### [expressão for variável in iterável if teste lógico]


```python
[i for i in [2,5,57,78,14,33] if i % 2 == 0]
```
