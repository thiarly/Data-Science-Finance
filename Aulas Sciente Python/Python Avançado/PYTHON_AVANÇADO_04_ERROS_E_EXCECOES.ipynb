### Os tipos de Erros

- *Erros de sintaxe:* o código não é válido em Python (geralmente fácil de corrigir)

- *Erros de tempo de execução:* o código sintaticamente é válido mas falha na execução


- *Erros semânticos:* erros na lógica: o código é executado sem problemas, mas o resultado não é o que você espera (geralmente muito difícil de rastrear e corrigir)

### Tratando exceções com ``try`` e ``except``

````python
try:
    print("tente isso:")
    x = 1 / 1 # Divisão por zero  
    print("consegui!")
    
except:
    print("algo ruim aconteceu!")
````
