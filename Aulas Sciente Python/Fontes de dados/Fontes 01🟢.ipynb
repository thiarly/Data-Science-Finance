```python
import yfinance as yf
```

```python
# Baixando cotações
ibov = yf.download('^BVSP')
ibov
```

```python
# Baixando cotações de ITUB4, entre 2015 e 2017
itub4 = yf.download('ITUB4.SA', start='2015-01-01', end='2017-12-31')
itub4
```

```python
# Baixando cotações dos últimos 12 meses de PETR4
petr4 = yf.download('ITUB4.SA', period='1y')
petr4

```

```python
# Baixando dados intraday, minuto a minuto de VALE3, dos últimos 5 pregões
# Valid intervals: 1m,2m,5m,15m,30,60m,90m,1h,1d,5d,1wk,1mo,3mo
# Valid periods: 1d,5d,1mo,3mo,6mo,1y,2y,5y,10y,ytd,max

vale3 = yf.download('VALE3.SA', interval='1m', period='5d')
vale3
```

```python
# Baixando várias empresas de uma vez. Método 1(Tickers separados por espaço)
jb = yf.download('JHSF3.SA BBAS3.SA')
jb
```

```python
# Baixando várias empresas de uma vez. Método 2 (Passando uma lista de tickers)
jb = yf.download(['JHSF3.SA', 'BBAS3.SA'])
jb
```

```python
# Alterando o método de agrupamento
# Group by 'ticker' or 'column' (default)

jb = yf.download('JHSF3.SA BBAS3.SA', group_by='ticker')
jb
```

```python
# Usando a classe Ticker
vale3 = yf.Ticker('VALE3.SA')
```

```python
# Dividendos e Splits (.actions)
vale3.actions
```

```python
# Informações da Empresa (.info)
vale3.info
```

```python
# Top Holders
vale3.institutional_holders
vale3.major_holders
```

```python
# Recomendações de Analistas
vale3.recommendations
```
