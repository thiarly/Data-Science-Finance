```python
import mplfinance as mpf
import yfinance as yf
```

```python
ibov = yf.download('^BVSP', period='3y')
```

```python
ibov.head()
```

## Tipos de Gráficos

### Linha

```python
mpf.plot(ibov, type='line')
```

### Renko

```python
mpf.plot(ibov,type='renko')
```

### Barras

```python
mpf.plot(ibov)
```

### Candlestick

```python
mpf.plot(ibov,type='candle')
```

## Estilos

```python
mpf.available_styles()
```

```python
for estilo in mpf.available_styles():
  mpf.plot(ibov, style=estilo, axtitle=estilo, type='candle', mav=(9,21,50), volume=True)
```

## Personalizando

```python
mpf.plot(ibov, 
         type='candle',
         style='charles',
         title='Índice Bovespa', 
         axtitle='Últimos 3 meses',
         ylabel='Pontos', 
         volume=True,
         ylabel_lower='volume',                  
         volume_exponent=False,
         datetime_format='%d/%m/%Y',
         xrotation=0,
         fontscale=0.8,
         #figsize=(10,5),
         hlines=[125000,95000],
         #vlines=['2021-03-03'],
         tight_layout=True,  
         #show_nontrading=True
         )
```
