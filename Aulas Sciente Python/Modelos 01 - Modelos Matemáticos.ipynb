## Conceitos de Probabilidade

**Experimento** é um processo que pode ser repetido e que gera uma série de resultados.

**Evento** é um conjunto de resultados

[link text](https://)image.png

**Espaço Amostral** é o conjunto de todos os resultados possíveis de nosso experimento aleatório

$S: \{ cara, coroa \}$

**Variável Aleatória** é uma função que atribui um número real a cada resultado do espaço amostral

$X$ é uma variável aleatória que assume valor $1$ quando ocorre $cara$ e $-1$ quando ocorre $coroa$

\begin{equation}
X =
\left \{
\begin{array}{cc}
1, & \text{se cara} \\
-1, & \text{se coroa} \\
\end{array}
\right.
\end{equation}

$$P(X = 1) = P(X = -1) = \frac{1}{2}$$

## Esperança

$E[X] = \sum x.p(x)$

$E[X] = 1.p(1) + (-1.p(-1))$

$E[X] = 1.\frac{1}{2} -1\frac{1}{2} = 0$
