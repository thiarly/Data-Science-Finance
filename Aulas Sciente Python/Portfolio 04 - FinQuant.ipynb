```python
from finquant.portfolio import build_portfolio
import matplotlib.pyplot as plt
```

```python
tickers = ['LEVE3.SA','BBAS3.SA','TAEE3.SA','BBSE3.SA','JHSF3.SA','FLRY3.SA',]
```

```python
pf = build_portfolio(names=tickers, data_api='yfinance')
```

```python
pf.risk_free_rate = 0.0275
```

```python
pf.properties()
```

```python
pf.mc_optimisation(50000)
```

```python
plt.figure(figsize=(18,8))
pf.mc_plot_results()
pf.ef_plot_efrontier()
pf.ef_plot_optimal_portfolios()
pf.plot_stocks()

plt.title('Fronteira Eficiente', fontsize=20)
plt.xlabel('Volatilidade', fontsize=14)
plt.title('Retorno Esperado', fontsize=12)
```
