```python
base_dados = ('/content/holding_homo.xlsm')
df = pd.read_excel(base_dados, sheet_name='carteira')
df
```

```python
# Substituir NaN por 0
df = df.fillna(0)

# Formatar as colunas como porcentagem (%)
df['Y1 2023'] = df['Y1 2023'].map('{:.2%}'.format)
df['Y1'] = df['Y1'].map('{:.2%}'.format)
df['Y2'] = df['Y2'].map('{:.2%}'.format)
df['POSIÇÃO'] = df['POSIÇÃO'].map('{:.2%}'.format)
df['DIFERENÇA'] = df['DIFERENÇA'].map('{:.2%}'.format)
df['OBJETIVO'] = df['OBJETIVO'].map('{:.2%}'.format)
df['FINANCEIRO'] = df['FINANCEIRO'].apply(lambda x: 'R$ {:,.2f}'.format(float(x)).replace(',', '.')).replace('.', ',')





# texto_lucro = f'R${lucro:_.2f}'
# texto_lucro = texto_lucro.replace('.',',').replace('_','.')
# print(f'O lucro foi de {texto_lucro}')




# Exibir o DataFrame formatado
df
```
