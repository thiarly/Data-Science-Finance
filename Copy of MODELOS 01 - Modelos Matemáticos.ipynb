## Conceitos de Probabilidade

**Experimento** é um processo que pode ser repetido e que gera uma série de resultados.

**Evento** é um conjunto de resultados

[link text](https://)image.png

**Espaço Amostral** é o conjunto de todos os resultados possíveis de nosso experimento aleatório

$S: \{ cara, coroa \}$


**Variável Aleatória** é uma função que atribui um número real a cada resultado do espaço amostral

$X$ é uma variável aleatória que assume valor $1$ quando ocorre $cara$ e $-1$ quando ocorre $coroa$

\begin{equation}
X =
\left \{
\begin{array}{cc}
1, & \text{se cara} \\
-1, & \text{se coroa} \\
\end{array}
\right.
\end{equation}


$$P(X = 1) = P(X = -1) = \frac{1}{2}$$

## Esperança

$E[X] = \sum x.p(x)$

$E[X] = 1.p(1) + (-1.p(-1))$

$E[X] = 1.\frac{1}{2} -1\frac{1}{2} = 0$

## Modelagem do Retorno

Como saber o retorno futuro de uma ação?

Não é possível!

Mas podemos recorrer a um Modelo Matemático para fazer uma estimativa

Vamos usar , então, um modelo estocástico (probabilístico):

Seja $X$ uma **variável aleatória** que representa o **retorno de um dado ativo** ao final de um período de tempo

O objetivo da modelagem é encontrar uma Distribuição de Probabilidade que se adeque aos dados

### Distribuições de Probabilidade

### O que caracteriza uma distribuição de Probabilidade?

#### 1) A função densidade de Probabilidade

#### 2) A função Probabilidade Acumulada

#### 2) Momentos

1º momento: a média

2º momento: variância

3º momento: assimetria (skewness)

4º momento: curtose (kurtosis)

### A Distribuição Normal

#### 1) A função Densidade de Probabilidade (pdf)

$$f(x) = \frac{1}{\sigma.\sqrt{2\pi}} e^{-\frac{1}{2}(\frac{x -\mu}{\sigma})^2}  $$

#### 2) Momentos

**1º Momento (média)**

$E[X] =  \mu$

**2º Momento (Variância)**

$ E[(X - \mu)^2] = \sigma^2$

**3º Momento (Assimetria / Skewness)**

$$\frac{E[(X - \mu)^3]}{\sigma^3}$$

Para Normal assimetria = 0

**4º Momento (Curtose / kurtosis)**

$$\frac{E[(X-\mu)^4]}{\sigma^4} $$
